```python
def viz_outliers(df, k, col_to_viz):
    for col in col_to_viz:
        if df[col].dtype != 'O':
            q1 = df[col].quantile(0.25)
            q3 = df[col].quantile(0.75)
            iqr = q3 - q1
            upper_bound =  q3 + k * iqr
            lower_bound =  q1 - k * iqr
            
            fig = plt.figure(figsize=(20,4))
            
            plt.suptitle(f'{col} Outlier Viz')
            #subplot 1
            plt.subplot(121)
            ax=sns.distplot(df[col])
            plt.axvline(lower_bound)
            plt.axvline(upper_bound)
            plt.title(f'Distribution of {col} with \n manual upper and lower bounds')
            
            
            #subplot 2
            plt.subplot(122)
            plt.title('Box Plot')
            sns.boxplot(df[col], color='coral', fliersize=k)
            plt.show()
viz_outliers(df, 1.25, outlier_cols)
```

## Hypotheses Testing

#### Hypothesis 1
Is there a relationship between properites with more than 2 bathrooms and logerror?

```python
train['bathroomcnt_bins'] = pd.cut(train.bathroomcnt, bins=[0, 2.5, 5])
```

```python
 sns.boxplot(x = train.bathroomcnt_bins, y=train.logerror)
```

Levene's Test: test for equal variance

$H_{0}$ : there is no difference in spending score variance between the two samples

$H_{a}$  : there is a diff in specnding score varaince between the two

```python
a = 0.05
stat, p = stats.levene(train[train.bathroomcnt <=2].logerror,
            train[train.bathroomcnt > 2].logerror)
if p > a:
    print('We accept null hypothesis, there is no difference in variance')
else:
    print('We reject the null hypothesis, there is a difference in variance')
    stat, p = stats.ttest_ind(train[train.bathroomcnt <= 2].logerror, 
                    train[train.bathroomcnt > 2].logerror,
                    equal_var=False,
    )
    print(f'stat: {stat}, p-value: {p}')
    if p < a:
        print(f'We reject the null hypothesis, there is a difference in logerror between properties with\n more than two baths and those with two baths or less.')
```

#### State Hypothesis:

$H_{0}$: population means of the logerror for properties built within each interval are all equal.

$H_{a}$: population means of the logerror for properties built within each interval are NOT all equal.

- All distributions are normal
- Samples are independent
- Use Levene test to check for variance

```python
# Set alpha
a= 0.05
```

```python
stat, p = stats.levene(train[(train.yearbuilt >= 1916) & (train.yearbuilt < 1944)].logerror, 
                         train[(train.yearbuilt >= 1944) & (train.yearbuilt < 1965)].logerror, 
                         train[(train.yearbuilt >= 1965) & (train.yearbuilt < 1994)].logerror, 
                         train[(train.yearbuilt >= 1995) & (train.yearbuilt < 2013)].logerror
                      )

if p > a:
    print('We accept null hypothesis, there is no difference in variance')
else:
    print('We reject the null hypothesis, there is a difference in variance')
    print('We reject the null hypothesis, we will use Kruskal-Wallis test')
```

```python
stats.kruskal(train[(train.yearbuilt >= 1916) & (train.yearbuilt < 1944)].logerror, 
                         train[(train.yearbuilt >= 1944) & (train.yearbuilt < 1965)].logerror, 
                         train[(train.yearbuilt >= 1965) & (train.yearbuilt < 1994)].logerror, 
                         train[(train.yearbuilt >= 1995) & (train.yearbuilt < 2013)].logerror
             )
```

**Conclusion**

Since p-value (0.00000281) < alpha (0.05), we reject the null hypothesis.

So the population means of the logerror for properties built within each interval are NOT all equal.

#### Hypothesis 3
Does logerror vary across properties in LA county vs properties out of LA county?

$H_{0}$ : there is no difference in spending score variance between the two samples

$H_{a}$  : there is a diff in specnding score varaince between the two

```python
train['LA_bin'] = [1 if county == 6111 else 0 for county in train.fips]
```

```python
train.LA_bin.value_counts()
```

```python
sns.boxplot(data=train, x=train.LA_bin, y='logerror')
```

```python
# We test variance with Levene Test to test variance
a = 0.05
stat, p = stats.levene(train[train.LA_bin == 1].logerror,
            train[train.LA_bin == 0].logerror)
if p > a:
    print('We accept null hypothesis, there is no difference in variance')
else:
    print('We reject the null hypothesis, there is a difference in variance')
    stat, p = stats.ttest_ind(train[train.LA_bin == 1].logerror,
                                    train[train.LA_bin == 0].logerror,
                                    equal_var=False
    )
    print(f'stat: {stat}, p-value: {p}')
    if p < a:
        print('We reject the null hypothesis, there is a difference in logerror between LA and non-LA properties.')
```
